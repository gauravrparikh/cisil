# Stacked Bar Plot

```python
reg = reg.rename(columns={"Study Card": "study_card", "Initial Load": "initial_load"})
reg
```

```python
reg['study_card'] = reg['study_card'].fillna(reg['initial_load'])
reg['study_card'] = reg.study_card.fillna('NaN')
reg.study_card.value_counts()
#reg.group_by('id')
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(reg.head(100))
```

```python
reg_first = reg[reg['stage'] == '1']

# id to study_card type mappings
reg_grouped_id_sc = reg_first[['id', 'study_card']].set_index('id')
reg_grouped_id_sc
```

```python
# merged df with study cards and stages
id_sc_stage = pd.merge(reg_grouped_id_sc, reg_grouped_id, left_index=True, right_index=True)
id_sc_stage.to_csv("id_studycard_stage.csv")
```

```python
# these are too slow: 
#id_sc_stage.plot.bar(stacked=True)
#id_sc_stage.plot.bar(rot=0)
```

```python
multi = id_sc_stage.groupby(['study_card', 'stage']).size().unstack(fill_value=0)
multi.columns = multi.columns.astype(str)
multi = multi.div(multi.sum(axis=1), axis=0)
```

```python
multi_t = multi.transpose()
multi_t.columns = multi_t.columns.astype(str)
multi_t = multi_t.rename(columns={'Phase1 Treatment': 'Phase1',
                       'Phase2 Treatment': 'Phase2',
                       'Subsidized Annual Pass': 'SAP'})

multi_t
```

```python
len(multi['1'])
len(multi['2'])
len(multi_t.columns)
```

```python
import matplotlib.pyplot as plt

labels = multi_t.columns
width = .5 # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, multi['1'], width, label='1')
ax.bar(labels, multi['2'], width, bottom=multi['1'], label='2')
ax.bar(labels, multi['3'], width, bottom=multi['1'], label='3')
ax.bar(labels, multi['4'], width, bottom=multi['1'], label='4')
ax.bar(labels, multi['5'], width, bottom=multi['1'], label='5')
ax.bar(labels, multi['6'], width, bottom=multi['1'], label='6')
ax.bar(labels, multi['8'], width, bottom=multi['1'], label='8')


ax.set_ylabel('Count')
ax.set_title('How many times people enrolled in ORCA-LIFT based on Treatment')
ax.legend()
plt.legend(title = "# times enrolled")
plt.xticks(rotation = 45)
plt.show()
#fig.savefig('per_treatment_reg.png', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
width = .5 # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, label='Men')
ax.bar(labels, women_means, width, bottom=men_means, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
