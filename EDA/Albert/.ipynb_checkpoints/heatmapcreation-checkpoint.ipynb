```python
"""
Cleans ACS csv and matches column names with column names.
"""

import pandas as pd
# read in acs_raw
acs_clean = pd.read_csv('../../../data/CLEAN_King_County_ACS_2019_tract.csv')
```

```python
acs_clean
```

```python
acs_clean['pct_black'] = acs_clean['estimate_total_africanamerican'] / acs_clean['estimate_total']
acs_clean['pct_white'] = acs_clean['estimate_total_white'] / acs_clean['estimate_total']
acs_clean['num_people_income_below_twohundredpct'] = acs_clean['less_than_10000'] + acs_clean['10000_to_14999'] + acs_clean['15000_to_19999'] + acs_clean['20000_to_24999']
acs_clean['pct_people_income_below_twohundredpct'] = acs_clean['num_people_income_below_twohundredpct'] / acs_clean['estimate_total']
ret = acs_clean[['census_tract', 'median_age_total', 'estimate_total', 'pct_black', 'pct_white', 'num_people_income_below_twohundredpct', 'pct_people_income_below_twohundredpct']]
ret
```

```python
ret.to_csv('../../../data/TOHEATMAP_King_County_ACS_2019_tract.csv')
```

1. num_people: `estimate_total`

2. pct_black: `estimate_total_africanamerican / estimate_total`

3. pct_white: `estimate_total_white / estimate_total`

4. num_people_income_below_twohundredpct: `less_than_10000 + 10000_to_14999 + 15000_to_19999 + 20000_to_24999`

5. pct_people_income_below_twohundredpct: `(less_than_10000 + 10000_to_14999 + 15000_to_19999 + 20000_to_24999) / estimate_total`

6. `gini_index`

7. `median_age_total`
